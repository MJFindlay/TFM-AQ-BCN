### Ip inferior 0.1mm ==> user value of 0.05 mm because this is non-zero
Now convert to numeric column changing commas to decimal points

```python
df_weather_airport['prec'] = pd.to_numeric(df_weather_airport['prec'].str.replace(',','.')).round(decimals=2)
```

### NaN null value to be treated as imputed by spline fit  https://www.geeksforgeeks.org/python-pandas-dataframe-interpolate/
